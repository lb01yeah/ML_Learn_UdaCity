### 策略policy
策略表示为希腊字母$\pi$  
- 决定性策略(deterministic policy)：is a mapping $\pi:\mathcal{S}-->\mathcal{A}$ ,输入的状态state,输出的是要采取的动作action.  
- 随机性策略stochastic policy: is a mapping $\pi : \mathcal{S} * \mathcal{A} -->[0,1]$ ,智能体能够随机的选择动作，$\pi \left ( a|s \right )=\mathbb{P}\left ( A_{t}=a|S_{t}=s \right )$, 其输入是：输入状态s和动作a, 输出返回的是智能体在状态s时采取动作a的概率。  


#### 什么样的措施来保证智能体的策略是最佳策略？
智能体与环境互动，通过该互动估算出最优动作值函数$\mathcal{q}^{*}$,然后智能体使用该值函数得出最优策略$\pi^{*}$。

### 状态值函数 state-value function
策略$\pi$的状态值函数(小写v表示，下标表示对应的策略)是环境状态的函数  

$\nu_{\pi}(s)=\mathbb{E}_\pi[G_{t}|S_{t}=s]$



### 贝尔曼方程
将MDP中任何状态的值表示为：即时奖励和下个状态的折扣值。


对于一般的MDP我们需要计算和的预期值，通常面对复杂的环境，我们无法确定即时奖励和下个状态。



在这个网格世界示例中，一旦智能体选择一个动作，

- 它始终沿着所选方向移动（而一般 MDP 则不同，智能体并非始终能够完全控制下个状态将是什么）
- 可以确切地预测奖励（而一般 MDP 则不同，奖励是从概率分布中随机抽取的）。

在这个简单示例中，我们发现任何状态的值可以计算为即时奖励和下个状态（折扣）值的和。

Alexis 提到，对于一般 MDP，我们需要使用期望值，因为通常即时奖励和下个状态无法准确地预测。的确，我们在之前的课程中发现，奖励和下个状态是根据 MDP 的一步动态特性选择的。在这种情况下，奖励$\mathcal{r}$ 和下个状态 $\mathcal{s}^{'}$ 是从（条件性）概率分布 p(s',r|s,a)中抽取的，贝尔曼预期方程（对于$v_\pi$）表示了任何状态 s 对于_预期即时奖励和下个状态的预期_值的值：

$v_\pi(s) = \text{} \mathbb{E}_\pi[R_{t+1} + \gamma v_\pi(S_{t+1})|S_t =s]$
