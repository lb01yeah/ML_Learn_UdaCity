## 特征缩放
特征缩放是执行非监督式学习时的一个重要预处理步骤，以允许同时分析多个特征。

实际样本的特征的数值，往往是有的区别很大，例如：体重差别很大，但是人的身高却差别相对小一点，当计算机进行实际计算是会产生误差。所以要对数值进行缩放。  
- 缩放后的特征在[0,1]之间，预估输出相对稳定
- 缺点：如果输入特征中有异常数据，缩放特征会比较棘手，X的最大值和最小值可能比较极端。

import numpy as np
def featureScaling(arr):

    arr = np.array(arr)
    max = np.max(arr)
    min = np.min(arr)
    new=[]
    for item in arr:
        if max != min :
            float (item)
            item =float(item-min)/(max-min)
            new.append(item)
        else:
            item=0.5
            new.append(item)
    return new
